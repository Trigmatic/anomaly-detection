# Exploratory Data Analysis

```python
%matplotlib inline

# Filter warnings
import warnings
warnings.filterwarnings("ignore")

# Data manipulation
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Set font scale and style
plt.rcParams.update({'font.size': 15})
```

```python
# Load Data
df = pd.read_csv('kdd99-unsupervised-ad.csv')
print("Data size:", df.shape) # print data size
df.columns = ['v_{}'.format(i) for i in range(30)] # column names
df.head()
```

```python
# Data Types
df.info()
```

```python
df.v_29 = df.v_29.map({'n': 'normal', 'o': 'attack'})
```

```python
label_pct = df.v_29.value_counts()*100/len(df)
label_ct =  df.v_29.value_counts()
pd.DataFrame({'labels': label_pct.index, 'count': label_ct.values, 'percentage': label_pct.values})
```

```python
# Descriptive Statistics
df.describe()
```

```python
# Near-zero variance variables
print(df.v_6.value_counts())
print(df.v_7.value_counts())
print(df.v_8.value_counts())
print(df.v_9.value_counts())
```

The variables above show two distinct values, indicating that they can be considered categorical.

```python
# Histogram
df.hist(bins = 50, figsize = (25,20))
plt.show()
```

The above histogram shows a lot of variables are peaked at zero, which either suggest the presence of outliers or a categorical variable encoded as integers
